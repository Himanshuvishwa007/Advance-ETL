# ETL Process: Extract, Transform, Load

ETL (Extract, Transform, Load) is a crucial data integration process used to gather data from multiple sources, process it to fit the desired structure, and finally load it into a data warehouse or database for analysis. 

1. **Extract**: Data is extracted from various sources like databases, APIs, or flat files.
2. **Transform**: The data is cleaned, transformed, and enriched by applying business logic, aggregations, and calculations.
3. **Load**: The transformed data is loaded into a target database, data warehouse, or another system for analysis and reporting.

In this notebook, we will walk through an ETL pipeline using data from a SQL Server database to demonstrate the process of transforming and enriching sales data.


## Data connection With SQL server 

###  **Import Libraries, Connect to Database, and Query Sales Data**

In this cell, we import the necessary libraries, connect to the SQL Server database, and run an SQL query to retrieve sales data from the `AdventureWorks2022` database.




```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

pd.set_option('display.float_format', '{:.2f}'.format)

# Connect to the database
engine = create_engine("mssql+pyodbc://HIMANSHU\\SQLEXPRESS/AdventureWorks2022?driver=ODBC+Driver+17+for+SQL+Server")

# Query for sales data
sales_query = """
SELECT 
    soh.SalesOrderID, 
    soh.OrderDate, 
    soh.CustomerID, 
    sod.ProductID, 
    sod.OrderQty, 
    sod.LineTotal, 
    p.Name AS ProductName, 
    pc.Name AS ProductCategory, 
    st.Name AS Territory
FROM Sales.SalesOrderHeader AS soh
INNER JOIN Sales.SalesOrderDetail AS sod ON soh.SalesOrderID = sod.SalesOrderID
INNER JOIN Production.Product AS p ON sod.ProductID = p.ProductID
INNER JOIN Production.ProductSubcategory AS psc ON p.ProductSubcategoryID = psc.ProductSubcategoryID
INNER JOIN Production.ProductCategory AS pc ON psc.ProductCategoryID = pc.ProductCategoryID
INNER JOIN Sales.SalesTerritory AS st ON soh.TerritoryID = st.TerritoryID
"""
sales_data = pd.read_sql(sales_query, engine)
sales_data.head()
```

## **Add Calculated Columns**

In this step, we add two new columns to the dataset: 
1. **TotalRevenue**, calculated by multiplying the order quantity and line total.
2. **OrderCategory**, which classifies the sales into 'Low Value', 'Medium Value', and 'High Value' based on the total revenue.


```python
# Add Total Revenue column
sales_data['TotalRevenue'] = sales_data['OrderQty'] * sales_data['LineTotal']

# Add an Order Category column based on revenue
sales_data['OrderCategory'] = pd.cut(
    sales_data['TotalRevenue'], 
    bins=[0, 500, 2000, 10000], 
    labels=['Low Value', 'Medium Value', 'High Value']
)
```

## **Aggregation and Trend Analysis**

In this step, two key analyses are performed:
1. **Total Sales by Product Category**: We aggregate the total revenue and order count per product category.
2. **Monthly Sales Trends**: We calculate the total revenue for each month to identify sales trends over time.



```python
# Total Sales by Product Category
category_sales = sales_data.groupby('ProductCategory').agg({
    'TotalRevenue': 'sum',
    'SalesOrderID': 'count'
}).rename(columns={'SalesOrderID': 'TotalOrders'})
print(category_sales)

# Monthly Sales Trends
sales_data['OrderMonth'] = pd.to_datetime(sales_data['OrderDate']).dt.to_period('M')
monthly_sales = sales_data.groupby('OrderMonth')['TotalRevenue'].sum()
print(monthly_sales)
```

### Visualizing Monthly Sales Trends
This cell generates a line chart to visualize the monthly sales revenue trend. The total revenue for each month is plotted, providing insights into sales performance over time.


```python
# Plot the Monthly Sales Trends
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='b', linewidth=2)

# Adding titles and labels
plt.title('Monthly Sales Revenue Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

### Enriching Sales Data with Customer Demographics
In this step, we join the sales data with customer demographic information, including the `StoreID`, to enrich the dataset with additional customer insights for further analysis.


```python
# Join with customer data to get customer demographics
customer_query = "SELECT CustomerID, StoreID FROM Sales.Customer"
customer_data = pd.read_sql(customer_query, engine)

# Merge with sales data
enriched_data = sales_data.merge(customer_data, on='CustomerID', how='left')
```

### Data Transformation and Formatting
In this step, we reset the index of the `category_sales` DataFrame and move the `ProductCategory` back to a regular column. Additionally, we round the `TotalRevenue` values to two decimal places for better readability and consistency.


```python
# Reset the index and move the 'ProductCategory' from the index to a regular column
category_sales.reset_index(inplace=True)
# Round the 'TotalRevenue' column to 2 decimal places
category_sales['TotalRevenue'] = category_sales['TotalRevenue'].round(2)
category_sales
```

### Save Aggregated Data to the Database
In this step, we save the aggregated `category_sales` DataFrame back into the SQL Server database as a new table called `CategorySalesSummary`. The `if_exists='replace'` argument ensures that if the table already exists, it will be replaced with the new data.


```python
# Save the aggregated data back to the database
category_sales.to_sql('CategorySalesSummary', engine, if_exists='replace', index=False)
```

### Convert 'OrderMonth' to String Format
This step converts the 'OrderMonth' column from a `Period` type to a string format (`YYYY-MM`). This transformation makes it easier to work with the data in string-based operations or when exporting to other formats.


```python
# Convert 'OrderMonth' from Period to string format (YYYY-MM)
enriched_data['OrderMonth'] = enriched_data['OrderMonth'].astype(str)
```

### Load the Transformed Data into SQL Server
This step loads the transformed `enriched_data` DataFrame into the SQL Server database, creating or replacing the `enriched_sales_data` table. The `index=False` argument ensures that the DataFrame index is not included in the table.


```python
# Now, load the data into SQL Server
enriched_data.to_sql('enriched_sales_data', engine, if_exists='replace', index=False)
print("Data loaded successfully!")
```

### Total Revenue by Product Category

This bar chart illustrates the total revenue generated by each product category. By visualizing the total revenue in a bar chart, we can quickly identify which categories contribute the most to overall sales. The chart includes:

- **X-axis**: Product categories
- **Y-axis**: Total revenue for each category
- The chart's color scheme is set to a light blue (`skyblue`) for a clean and professional look.

The chart helps in understanding the distribution of revenue across different product categories, which is useful for sales strategy and decision-making.


```python
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', x='ProductCategory', y='TotalRevenue', color='skyblue', legend=False)
plt.title('Total Revenue by Product Category', fontsize=14)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Sales Order Distribution by Product Category

This pie chart displays the distribution of sales orders across different product categories. It helps in understanding the proportion of total sales orders attributed to each category. Key features of the chart:

- **Slices**: Each slice represents a product category, showing its share of total sales orders.
- **Labels**: The product categories are labeled directly on the chart.
- **Percentage**: The percentage of total orders for each product category is shown on each slice.

This visualization provides insights into which product categories are generating the most orders, which can inform inventory management and marketing strategies.


```python
category_sales.plot(kind='pie', y='TotalOrders', labels=category_sales['ProductCategory'], autopct='%1.1f%%', figsize=(8, 8))
plt.title('Sales Order Distribution by Product Category', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()
```
